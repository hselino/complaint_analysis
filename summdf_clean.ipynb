```python
from datasets import load_dataset

splits = {'train': 'train.csv', 'validation': 'validation.csv', 'test': 'test.csv'}
df = pd.read_csv("hf://datasets/batubayk/TR-News/" + splits["train"])
```

```python
print("Initial data size:", len(df))
df.drop_duplicates(subset=["abstract", "content", "title"], inplace=True)
df.dropna(subset=["abstract", "content", "title"], inplace=True)
print("Data size after removing duplicates/NaN:", len(df))
```

```python
columns = ["abstract", "content", "title"]
```

```python
df2 = df[columns].copy()
```

```python
df2
```

```python
def clean_text(text):
    text = text.lower()
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove HTML entities like &nbsp;
    text = re.sub(r"&\w+;", " ", text)

    # Remove all punctuation (including " and ')
    text = re.sub(r'[^\w\s]', '', text)

    # Normalize spaces (remove multiple spaces)
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df2["abstract"] = df2["abstract"].apply(clean_text)
df2["content"] = df2["content"].apply(clean_text)
df2["title"] = df2["title"].apply(clean_text)
```

```python
df["abstract"] = df["abstract"].apply(clean_text)
df["content"] = df["content"].apply(clean_text)
df["title"] = df["title"].apply(clean_text)
```

```python
URL_PATTERN = re.compile(r'https?://\S+|www\.\S+')
EMAIL_PATTERN = re.compile(r'\S+@\S+')
def remove_noise(text):
    text = URL_PATTERN.sub('', text)
    text = EMAIL_PATTERN.sub('', text)
    return text
```

```python
df2["abstract"] = df2["abstract"].apply(remove_noise)
df2["content"] = df2["content"].apply(remove_noise)
df2["title"] = df2["title"].apply(remove_noise)
```

```python
df2.to_csv("summ.csv", index=False)
```
